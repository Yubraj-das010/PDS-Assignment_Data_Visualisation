- ### 2.3: The General Audience and the Critics    

```python
# Change the scale of MetaCritic
class preparation():
    def critic(self):
        movies_final.MetaCritic = movies.MetaCritic/10
        movies_final["Avg_rating"] = (movies_final["MetaCritic"] + movies_final["IMDb_rating"])/2
        return movies_final["Avg_rating"]
```

```python
prep = preparation()
prep.critic()
```

```python
#Retain only the movies in which the absolute difference(using abs() function) 
#between the IMDb_rating and Metacritic columns is less than 0.5
#i have saved this data data to "movies1" so as to keep "movies" data set out of filteration
class Final():
        @staticmethod
        def movie():
            movies1 =  movies_final[abs(movies_final["MetaCritic"] - movies_final["IMDb_rating"])<0.5]
            movies1 = movies1.sort_values(["Avg_rating"],ascending = False)
            UniversalAcclaim =  movies1[movies1["Avg_rating"]>=8]
            UniversalAcclaim = UniversalAcclaim.sort_values(["Avg_rating"],ascending = False)
            return UniversalAcclaim
```

## Task 3 : Demographic analysis

-  ###  Subtask 3.1 Combine the Dataframe by Genres

```python
# Create the dataframe df_by_genre
cols = list(movies_final.columns.values)
df_by_genre = movies_final[cols[11:14] +cols[16:-5]]
df_by_genre["cnt"] = 1
df_by_g1 = df_by_genre.groupby("genre_1").sum()
df_by_g2 = df_by_genre.groupby("genre_2").sum()
df_by_g3 = df_by_genre.groupby("genre_3").sum()
df_by_g1andg2 = df_by_g1.add(df_by_g2,fill_value = 0.0)
df_add = df_by_g1andg2.add(df_by_g3,fill_value = 0.0)
        
genre_top10 = df_add[df_add["cnt"]>=10]
genre_top10.iloc[:,:-1] = genre_top10.iloc[:,:-1].div(genre_top10.cnt.values,axis=0)
genre_top10 =  round(genre_top10.iloc[:,:],2)
genre_top10 = genre_top10[genre_top10.iloc[:,:].columns].astype(int)


genre_top10.sort_values("CVotes1000", ascending =False,inplace=True)
genre_top10

# Creating IFUS column
movies_final["IFUS"] = movies_final["Country"].apply(lambda x:"USA" if x=="USA" else "non-USA")

class genre:
    def genreDF(self):
        plt.figure(figsize=[10,5])
        sns.barplot(genre_top10.index,genre_top10.cnt).set_title('Count of Movies by Genre')
        plt.show() 
    def heatmap_fav(self):
        fig, (axes)= plt.subplots(1,2,figsize = [15,5])
        sns.heatmap(genre_top10[['CVotesU18M','CVotes1829M','CVotes3044M','CVotes45AM']],cmap="Greens",annot = True, fmt="1g", ax = axes[0]).set_title('Male votes for a Genre')
        sns.heatmap(genre_top10[['CVotesU18F','CVotes1829F','CVotes3044F','CVotes45AF']],cmap="Greens",annot = True, fmt="1g", ax = axes[1]).set_title('Female votes for a Genre')
        plt.show()
    def heatmap_rate(self):
        fig, (axes)= plt.subplots(1,2,figsize = [15,5])

        sns.heatmap(genre_top10[['VotesU18M','Votes1829M','Votes3044M','Votes45AM']],cmap="Greens",annot = True, fmt="1g", ax = axes[0]).set_title('Rating by Males for a Genre')
        sns.heatmap(genre_top10[['VotesU18F','Votes1829F','Votes3044F','Votes45AF']],cmap="Greens",annot = True, fmt="1g", ax = axes[1]).set_title('Rating by Females for a Genre')
        plt.show()

genredf = genre()
```

#### Using concept of Inheritance

```python
class graph:
    def hist(self):
        plt.figure(figsize=[10,5])
        sns.barplot(genre_top10.index,genre_top10.CVotes1000).set_title('Votes for each genre')
        plt.show()
    def boxplot(self):
        sns.boxplot(movies_final.IFUS, movies_final.CVotesUS)
        plt.figure(figsize=[10,5])
        plt.show()
```

```python
class All(graph):
    def __init__(self):
        graph.hist(self)
        graph.boxplot(self)
        print("All")
```
